```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("car.csv")
```

```python
df.groupby("Make")["MSRP"].max()
```

```python
df.groupby(["Make","Vehicle Style"])["MSRP"].max()
```

```python
convertible_car = df[df["Vehicle Style"]== "Convertible" ]
most_exp = convertible_car.groupby("Make")["MSRP"].max()
print(most_exp)

```

```python
# df[df["Transmission Type"] == ["AUTOMATIC"] ]
df[df["Transmission Type"] == "AUTOMATIC"]
```

```python
primium_car = df[df["Engine Fuel Type"] == "premium unleaded (required)"]
most_expensive_premium_cars = primium_car.groupby("Make")["MSRP"].max()
print(most_expensive_premium_cars)
```

```python
each_car_avg = df.groupby("Make")["MSRP"].mean()
print(each_car_avg)
```
